##Cleaning of data

```python
dataset['Price'] = dataset['Price'].str.replace(',','').astype(int)
```

```python
dataset['Kms_driven'] = dataset['Kms_driven'].str.split(' ').str.get(0)
```

```python
dataset['Kms_driven'] = dataset['Kms_driven'].str.replace(',','').astype(int)
```

```python
dataset['Name'] =  dataset['Name'].str.split(' ').str.slice(0,3).str.join(' ')
```

##**Model**

```python
x = dataset.drop(columns='Price')
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
he = OneHotEncoder()
he.fit(x[['Name','Company','Fuel_type']])
OneHotEncoder()
```

```python
he.categories_
```

```python
column_tranform = make_column_transformer((OneHotEncoder(categories= he.categories_),['Name','Company','Fuel_type']),remainder= 'passthrough')
```

```python
model= LinearRegression()
```

```python
pipe = make_pipeline( column_tranform ,model)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict
```
